```python
import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.model_selection import train_test_split

from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split, cross_val_score
from collections import Counter
import pandas as pd
from nltk.corpus import stopwords
import string
from sklearn.decomposition import PCA







```

```python
def load_data(data_frame):
    reviews = []
    labels = []
    for index, row in data_frame.iterrows():
        label, review = row[0].split('\t', 1)  # Split each row based on the tab character
        labels.append(label.strip())  # Strip any leading or trailing whitespace
        reviews.append(review.strip())  # Strip any leading or trailing whitespace
    return reviews, labels
```

```python
def load_test_data(data_frame):
    reviews = []
    for row in data_frame.itertuples(index=False):
        review = row[0]  #   review text is in the first column
        reviews.append(review)
    return reviews
```

```python


nltk.download('punkt')
nltk.download('stopwords')

#  stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Convert tokens to lowercase
    tokens = [token.lower() for token in tokens]

    # Remove punctuation and stopwords
    tokens = [token for token in tokens if token not in punctuation and token not in stop_words]

    return tokens

def preprocess_texts_array(texts):
    preprocessed_texts = []
    for text in texts:
        preprocessed_text = preprocess_text(text)
        preprocessed_texts.append(preprocessed_text)
    return preprocessed_texts
```

```python

from sklearn.decomposition import PCA


from sklearn.feature_extraction.text import TfidfVectorizer






def cosine_similarity__1(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    magnitude_vec1 = np.linalg.norm(vec1)
    magnitude_vec2 = np.linalg.norm(vec2)

    if magnitude_vec1 == 0 or magnitude_vec2 == 0:
        return 0

    cosine_sim = dot_product / (magnitude_vec1 * magnitude_vec2)

    return cosine_sim





def euclidean_distance(vec1, vec2):
    squared_diff = (vec1 - vec2) ** 2
    sum_squared_diff = np.sum(squared_diff)
    euclidean_dist = np.sqrt(sum_squared_diff)
    return euclidean_dist

class KNNClassifier:
    def __init__(self, k):
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        predictions = []
        for x in X_test:
            # Calculate cosine similarity between x and all examples in the training set
            similarities = cosine_similarity([x], self.X_train)[0]
            # Sort by similarity and return indices of the first k neighbors
            k_indices = np.argsort(similarities)[-self.k:]
            # Extract the labels of the k nearest neighbor training samples
            k_nearest_labels = [self.y_train[i] for i in k_indices]
            # Return the most common class label
            prediction = max(set(k_nearest_labels), key=k_nearest_labels.count)
            predictions.append(prediction)
        return predictions
```

```python
def k_fold_cross_validation(X, y, k):

    fold_size =3

    accuracies = []
    for i in range(k):
        # Split data into training and validation sets
        validation_X = X[i*fold_size:(i+1)*fold_size]
        validation_y = y[i*fold_size:(i+1)*fold_size]
        train_X = np.concatenate([X[:i*fold_size], X[(i+1)*fold_size:]])
        train_y = np.concatenate([y[:i*fold_size], y[(i+1)*fold_size:]])

        # Initialize KNN classifier
        knn = KNNClassifier(k=i)  # Adjust k value as needed

        # Train the classifier on the training set
        knn.fit(train_X, train_y)

        # Evaluate the classifier on the validation set
        predictions = knn.predict(validation_X)
        accuracy = np.mean(predictions == validation_y)
        accuracies.append(accuracy)

    # Calculate the mean accuracy across all folds
    mean_accuracy = np.mean(accuracies)

    return mean_accuracy



```

```python
# Adjust the number of components as needed
pca = PCA(n_components=200)


```

```python

tfidf_train_data_pca = pca.fit_transform(tfidf_train_data.toarray())

# Transform the TF-IDF test data using the trained PCA
tfidf_test_data_pca = pca.transform(tfidf_test_data.toarray())

```

```python
k =100  # Number of folds
mean_accuracy = k_fold_cross_validation(tfidf_train_data_pca, x_train_labels,k)
print("Mean Accuracy:", mean_accuracy)

```
